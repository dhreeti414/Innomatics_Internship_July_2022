<h1><span style="color:DarkOliveGreen">MEAN</span></h1>

<p><span style="color:blue">The Mean value of a dataset is the average value i.e. a number around which a whole data is spread out. All values used in calculating the average are weighted equally when defining the Mean.</span></p>

<h3>For Example:</h3>

<h3>Formulae:</h3>

<h2>Population Mean: $\mu$  $$ \mu = \sum\frac{x}{N} $$ </h2>

where, x= total of the numbers<br>
       N= total numbers

<h2>Sample Mean: $\bar x $ $$\bar x = \sum\frac{x}{n} $$ </h2>

where, x=total of the numbers<br> 
       n= total numbers

<h2>Geometric Mean: $$ GM(x) = \sqrt[n]{x_1\times x_2\times ...\times x_n} $$ </h2>

<h2>Harmonic Mean: $$ HM(x) = \frac{n}{\frac{1}{x_1}+\frac{1}{x_2}+...+\frac{1}{x_n}}$$ </h2>

<h1><span style="color:DarkOliveGreen">VARIANCE</span></h1>

<span style='color:blue'>The term variance refers to a statistical measurement of the spread between numbers in a data set. More specifically, variance measures how far each number in the set is from the mean (average), and thus from every other number in the set.</span>

<h3>For Example:</h3>

<h3>Formulae:</h3>

<h2>Population Variance: $$ \sigma^2 = \frac{\sum\,(X_i - \mu)^2}{N} $$ </h2>

where,<br>
$\sigma^2$ = population variance<br>
$\sum$ = sum <br>
Χ = each value<br>
$\mu$ = population mean<br>
Ν = number of values in the population<br>

<h2>Sample Variance : $$ \sigma^2 = \frac{\sum\,(X_i - \bar{x})^2}{n-1} $$ </h2>

where,<br>
$\sigma^2$ = sample variance<br>
$\sum$ = sum <br>
Χ = each value<br>
$\bar{x}$ = sample mean<br>
n = number of values in the sample

<h3>Code Examples:</h3>

<h1><span style="color:DarkOliveGreen">STANDARD DEVIATION</span></h1>

<span style='color:blue'>The standard deviation is the most common way of quantifying the spread of a data. It is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance. The standard deviation is calculated as the square root of variance by determining each data point's deviation relative to the mean.</span>

<h3>Formulae:</h3>

<h2>Population Standard Deviation: $$ \sigma = \sqrt{\frac{\sum\,(X_i - \mu)^2}{N}} $$ </h2>

where,<br>
$\sigma$=Standard Deviation<br>
X=Terms given in the data<br>
N=Total number of terms<br>
$\mu$=Mean of the population

<h2>Sample Standard Deviation : $$  \sigma = \sqrt{\frac{\sum\,(X_i - \bar{x})^2}{N-1}} $$ </h2>

where,<br>
$\sigma$=Standard Deviation<br>
X=Terms given in the data<br>
N=Total number of terms<br>
$\bar x $= Mean of the data

<h3>Code Examples:</h3>

<h1 style="color:DarkOliveGreen">CORRELATION</h1>

<span style="color:blue">Correlation is a statistical measure (expressed as a number) that describes the size and direction of a relationship between two or more variables. It measures the relationship between two variables.</span>

<h3>For Example:</h3>

<h2>Types of Correlation</h2>

<h3>Pearson Correlation</h3>

<span style="color:gray">Pearson correlation is also known as the Pearson product-moment correlation coefficient and is a normalized measurement of the covariance. It also measures the linear relationship between two variables and fails to capture the non-linear relationship of two variables. Pearson correlation assumes that both variables are normally distributed. It can be used for nominal variables or continuous variables.</span>

<h3>Formula:</h3>

<h2>Pearson Correlation Coefficient: r  $$ r = \frac{\sum(x_i-\bar x)(y_i-\bar y)}{\sqrt{\sum(x_i-\bar x)^2\sum(y_i-\bar y)^2}} $$ </h2>

where,<br>
$x_i$=x variable sample<br>
$y_i$=y variable sample<br>
$\bar x$=mean of values in x variable<br>
$\bar y$=mean of values in y variable<br>

<h3>Example</h3>

```python
from scipy.stats import pearsonr
```

```python
a = np.random.normal(size=10)
b = np.random.normal(size=10)
```

```python
pearsonr(a,b)
```

<h3>Spearman’s Rank Correlation</h3>

<span style="color:gray">It is a nonparametric(no prior assumptions about distribution) measure for calculating correlation coefficient that is used for ordinal variables or continuous variables. Spearman’s rank correlation can capture both linear or non-linear relationships.</span>

<h3>Formula:</h3>

<h2>Spearman’s Rank Correlation Coefficient:$\rho$ $$\rho = 1- \frac{6\sum d_i ^2}{n(n^2 -1)}$$</h2>

where,<br>
$d_i$=difference between the two ranks of each observation<br>
n= number of observations

<h1 style="color:DarkOliveGreen">NORMAL DISTRIBUTION</h1>

<span style="color:blue">Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.
In graphical form, the normal distribution appears as a "bell curve".</span>

<h3>For Example:</h3>

<h3>Formula:</h3>

<h2>Normal Distribution:</h2>
<p> the values of the mean (μ) and standard deviation(σ)are necessary</p>

<h2>$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{2}\right) $$ </h2>

where,<br>
x = value of the variable or data being examined<br>
f(x)= the probability function

<h3>Code Example:</h3>

```python
from scipy.stats import norm
cdf_upper_limit = norm(loc = 5.3 , scale = 1).cdf(6.5)
cdf_lower_limit = norm(loc = 5.3 , scale = 1).cdf(4.5)
 
prob = cdf_upper_limit - cdf_lower_limit
print(prob)
```

<h1 style="color:DarkOliveGreen">FEATURE OF NORMAL DISTRIBUTION</h1>

<span style="color:purple">1. It is symmetric</span><br>
A normal distribution comes with a perfectly symmetrical shape. This means that the distribution curve can be divided in the middle to produce two equal halves. The symmetric shape occurs when one-half of the observations fall on each side of the curve.
<br><br>
<span style="color:purple">2. The mean, median, and mode are equal</span><br>
The middle point of a normal distribution is the point with the maximum frequency, which means that it possesses the most observations of the variable. The midpoint is also the point where these three measures fall. The measures are usually equal in a perfectly (normal) distribution.
<br><br>
<span style="color:purple">3. Empirical rule</span><br>
In normally distributed data, there is a constant proportion of distance lying under the curve between the mean and specific number of standard deviations from the mean. For example, 68.25% of all cases fall within +/- one standard deviation from the mean. 95% of all cases fall within +/- two standard deviations from the mean, while 99% of all cases fall within +/- three standard deviations from the mean.
<br><br>
<span style="color:purple">4. Skewness and kurtosis</span><br>
Skewness and kurtosis are coefficients that measure how different a distribution is from a normal distribution. Skewness measures the symmetry of a normal distribution while kurtosis measures the thickness of the tail ends relative to the tails of a normal distribution.<br>

<h1 style="color:DarkOliveGreen">POSITIVELY SKEWED AND NEGATIVELY SKEWED NORMAL DISTRIBUTION</h1>

<span style="color:blue">A left-skewed distribution has a long left tail. Left-skewed distributions are also called<span style="color:red"> negatively-skewed distributions</span>. That’s because there is a long tail in the negative direction on the number line. The mean is also to the left of the peak.</span>

<span style="color:blue">A right-skewed distribution has a long right tail. Right-skewed distributions are also called <span style="color:red">positive-skew distributions</span>. That’s because there is a long tail in the positive direction on the number line. The mean is also to the right of the peak.</span>

<h3>Formulae:</h3>

<h2>Pearson's first skewness coefficient (mode skewness):<br><br>$$ \frac{Mean - Mode}{StandardDeviation}$$</h2>

<h2>Pearson's second skewness coefficient (median skewness):<br><br> $$ \frac{3(Mean - Median)}{StandardDeviation}$$

<h1 style="color:DarkOliveGreen">Box Cox Transformation</h1>

<span style="color:blue">The Box-Cox transformation transforms our data so that it closely resembles a normal distribution.<br>
In many statistical techniques, we assume that the errors are normally distributed. This assumption allows us to construct confidence intervals and conduct hypothesis tests. By transforming your target variable, we can (hopefully) normalize our errors (if they are not already normal).<br>
Additionally, transforming our variables can improve the predictive power of our models because transformations can cut away white noise.</span>

<h3>For Example:</h3>

<h3>Mathematical Formula:</h3>

<h2> The transformation equation is as follows:<br><br>
$$ y(\lambda) = \left\{\begin{array}\\ \left(y^ \lambda -1\right)/\lambda & \mbox{if}\ \lambda \neq {0}\\
\log {y} & \mbox{if}\ \lambda = 0 \end{array}\right. $$ </h2>

<h3>Code Examples:</h3>

```python
plt.figure(figsize = (8, 8))
data = df['Mthly_HH_Income'] 
sb.histplot(data,kde=True)
plt.show()
```
