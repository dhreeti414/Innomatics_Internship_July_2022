<h1>Introduction of Dataset</h1>

Diamond is one of the strongest and the most valuable substances produced naturally as a form of carbon. However, unlike gold and silver, determining the price of a diamond is very complex because many features are to be considered for determining its price.

The value of diamonds depends upon their structure, cut, inclusions (impurity), carats, and many other features. The uses of diamonds are many, such as in industries, as they are effective in cutting, polishing, and drilling. Since diamonds are extremely valuable, they have been traded across different countries for centuries now and this trade only increases with time.

The given Diamond dataset have 10 features and approx 54000 rows or instances. It has the features such as price, carat, cut, clarity, color, x(length), y(width) and z(depth) in order to predict the price of the Diamond based on the given features

<h2><i>Experiment Tracking using MLFlow</i></h2>

```python
import mlflow
```

```python
mlflow.set_tracking_uri("sqlite:///diamond.db")
```

```python
mlflow.set_experiment("Diamond Price Prediction")
```

```python
from pickle import dump

dump(scaler, open('models/standard_scaler.pkl', 'wb'))
```

<h2><i>Experiment Runs</i></h2>

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

<h3 style="color:DarkSlateBlue">Experiment 1 - Training Linear Regression</h3>

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "Dhreeti")
    mlflow.set_tag("algo", "LR")
    mlflow.log_param("data-path", "diamonds.csv")
    linear_regressor = LinearRegression()
    linear_regressor.fit(X_train_new, y_train)
    y_test_pred = linear_regressor.predict(X_test_new)
    r2score_acc_lr = metrics.r2_score(y_test, y_test_pred)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
    mlflow.log_metric("accuracy", r2score_acc_lr)
    mlflow.log_metric("RMSE", rmse)
    mlflow.sklearn.log_model(linear_regressor, artifact_path = "models")
    mlflow.log_artifact("models/standard_scaler.pkl")
```

<h3 style="color:DarkSlateBlue">Experiment 2 - Training K Nearest Neighbour (KNN) Regressor</h3>

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "Dhreeti")
    mlflow.set_tag("algo", "KNN")
    mlflow.log_param("data-path", "diamonds.csv")
    k = 5
    mlflow.log_param("n_neighbors", k)
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(X_train_new, y_train)
    y_test_pred = knn_regressor.predict(X_test_new)
    r2score_acc_knn = metrics.r2_score(y_test, y_test_pred)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
    mlflow.log_metric("accuracy", r2score_acc_knn)
    mlflow.log_metric("RMSE", rmse)
    mlflow.sklearn.log_model(knn_regressor, artifact_path="models")
    mlflow.log_artifact("models/standard_scaler.pkl")
```

<h3 style="color:DarkSlateBlue">Experiment 3 - Training Random Forest Regressor</h3>

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "Dhreeti")
    mlflow.set_tag("algo", "Random Forest Regression")
    mlflow.log_param("data-path", "diamonds.csv")
    r = 50
    rf_regressor = RandomForestRegressor(n_estimators = r)
    rf_regressor.fit(X_train_new, y_train)
    y_test_pred = rf_regressor.predict(X_test_new)
    r2score_acc_rf = metrics.r2_score(y_test, y_test_pred)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
    mlflow.log_param("n_estimators", r)
    mlflow.log_metric("accuracy", r2score_acc_rf)
    mlflow.log_metric("RMSE", rmse)
    mlflow.sklearn.log_model(rf_regressor, artifact_path = "models")
    mlflow.log_artifact("models/standard_scaler.pkl")
```

<h3 style="color:DarkSlateBlue">Experiment 4 - Training Decision Tree Regressor</h3>

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "Dhreeti")
    mlflow.set_tag("algo", "Decision Tree Regression")
    mlflow.log_param("data-path", "diamonds.csv")
    d = None
    dt_regressor = DecisionTreeRegressor(max_depth = d)
    dt_regressor.fit(X_train_new, y_train)
    y_test_pred = dt_regressor.predict(X_test_new)
    r2score_acc_dt = metrics.r2_score(y_test, y_test_pred)
    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
    mlflow.log_param("max_depth", d)
    mlflow.log_metric("accuracy", r2score_acc_dt)
    mlflow.log_metric("RMSE", rmse)
    mlflow.sklearn.log_model(dt_regressor, artifact_path = "models")
    mlflow.log_artifact("models/standard_scaler.pkl")
```

<h3 style="color:DarkSlateBlue">Experiment 5 - Training KNN Regressor with Hyperparameter Tuning</h3>

```python
from sklearn.model_selection import GridSearchCV
mlflow.sklearn.autolog(max_tuning_runs = None)
```

```python
with mlflow.start_run():
    tuned_parameters = [{'n_neighbors': [i for i in range(1, 51)], 'p': [1, 2]}]
    gs = GridSearchCV(
        estimator = KNeighborsRegressor(),
        param_grid = tuned_parameters,
        scoring = 'neg_mean_squared_error',
        cv = 5,
        return_train_score = True,
        verbose = 1
    )
    gs.fit(X_train_new, y_train)
    mlflow.sklearn.autolog(disable = True)
```
